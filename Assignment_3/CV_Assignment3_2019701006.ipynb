# Experimenting Parameters :

```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
import os 
import sys
import random
import warnings
from grabcut import GrabCut
warnings.filterwarnings('ignore')
```

```python
image_dir = 'input_data/images/'
bbox_dir = 'input_data/bboxes/'
img_list = []
for filename in os.listdir(image_dir):
    im = cv.imread(os.path.join(image_dir,filename))
    bbox_file = filename.split('.')[0] + '.txt'
    with open(os.path.join(bbox_dir,bbox_file)) as f:
        mask_coord = [int(x) for x in f.read().split()]
    img_list.append((im,mask_coord))
print("Total Images:",len(img_list))
```

##  Number of iterations of GMM 

```python
k = 5
#images_list = random.choices(img_list, k = k)
for img,dimen in img_list:    
    i = 1
    img = img.astype('uint8')
    x1,y1,x2,y2 = dimen
    rectangle = (x1,y1,x2-x1,y2-y1)
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(k,5,i)
    ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    ax.axis('off')
    ax.set_title('Original Image')
    # Number of itersaions
    for iterations in [1,10,30]:            
        # Define mask
        mask = np.zeros(img.shape[:2],np.uint8)
        bgd = np.zeros((1,65),np.float64)
        fgd = np.zeros((1,65),np.float64)
        mask, bgd, fgd = cv.grabCut(img, mask, rectangle, bgd, fgd, iterations, cv.GC_INIT_WITH_RECT)
        mask1 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
        segmented_img = img*mask1[:,:,np.newaxis]
        i += 1
        ax = fig.add_subplot(k,5,i)
        ax.imshow(cv.cvtColor(segmented_img, cv.COLOR_BGR2RGB))
        ax.axis('off')
        ax.set_title('Iteration : '+str(iterations))
        
    plt.show()
```

We can see how by increasing the number of interations of the GMM we are better able to approximate the mincut

##  Number of mixture components in GMM

```python
iterations = 1
k = 5
#images_list = random.choices(img_list, k = 5)
for img,coord in img_list:    
    i = 1
    img = img.astype('uint8')
    x1,y1,x2,y2 = coord
    rectangle = (x1,y1,x2-x1,y2-y1)
    output = np.zeros(img.shape, np.uint8)
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(k,4,i)
    ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    ax.axis('off')
    ax.set_title('Original Image')
    for n_components in [1,5,25]:            
        mask = np.zeros(img.shape[:2],np.uint8)
        mask[y1:y2,x1:x2] = 2        

        gc = GrabCut(img, mask, rectangle, gmm_components=n_components)
        gc.run()
        mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
        seg_img = img*mask2[:,:,np.newaxis]
        i += 1
        ax = fig.add_subplot(k,4,i)
        ax.imshow(cv.cvtColor(seg_img, cv.COLOR_BGR2RGB))
        ax.axis('off')
        ax.set_title('GMM components : '+str(n_components))
        
    plt.show()
```

Here we can notice we get good result on 5 components. Number of components to use can be very vague can depends on the image and the forground

## Diﬀerent color spaces or ways of representing pixels

```python
iterations = 4
cname = {
    4: 'RGB',
    44: 'LAB',
    36: 'YCrCb'
}
k = 5
for img,coord in img_list:  
    i = 1
    img = img.astype('uint8')
    x1,y1,x2,y2 = coord
    rectangle = (x1,y1,x2-x1,y2-y1)
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(k,4,i)
    ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    ax.axis('off')
    ax.set_title('Original Image')
    # Color spaces 
    for color_space in [cv.COLOR_BGR2RGB, cv.COLOR_BGR2LAB, cv.COLOR_BGR2YCR_CB]:            
        mask = np.zeros(img.shape[:2],np.uint8)
        bgd = np.zeros((1,65),np.float64)
        fgd = np.zeros((1,65),np.float64)
        
        mask, bgd, fgd = cv.grabCut(cv.cvtColor(img, color_space),mask,rectangle,bgd,fgd,iterations,cv.GC_INIT_WITH_RECT)
        mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
        segmented_image = img*mask2[:,:,np.newaxis]
        i += 1
        ax = fig.add_subplot(k,4,i)
        ax.imshow(cv.cvtColor(segmented_image, cv.COLOR_BGR2RGB))
        ax.axis('off')
        ax.set_title('Result on : ' + cname[color_space])
        
    plt.show()
```

The color space seems to have less effect on the algorithm
